```python
import pandas as pd
```

```python
cluster_root_path = './data/training_data/blog/'
excel_root_path = './data/training_data/'

excel_names = ['1128.xlsx','1205.xlsx','1212.xlsx']
excel_paths = []
for name in excel_names:
    excel_paths.append(excel_root_path + name)
# print(excel_paths)

cluster_names = ['贬义/','中性/','褒义/']
cluster_paths = []
for name in cluster_name:
    cluster_paths.append(cluster_root_path + name)
# print(cluster_paths[1])
```

```python
for xls_path in excel_paths:
    sheet = pd.read_excel(xls_path, usecols=[2,3], header = None)
    sheet = sheet.drop([0])
    print(sheet)
    for index,row in sheet.iterrows():
        num = xls_path[-9:-5]
        name = str(index)
        txt_name = num+'_'+name+'.txt'
        
        flag = row[0]
        txt_root = cluster_paths[flag+1]
        txt_path = txt_root + txt_name
        # print(txt_path)
        file = open(txt_path, 'w')
        text = row[1]
        file.write(text)
        
        
    
    
    
    
```
