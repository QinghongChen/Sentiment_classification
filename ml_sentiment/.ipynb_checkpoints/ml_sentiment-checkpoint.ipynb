```python
import numpy as np
import pandas as pd
from pyhanlp import *
```

```python
test_data_path = './data/test_data/yearly_movies.xls'
#test_data_path = './data/test_data/Hotels.xls'
```

```python
#朴素贝叶斯
NaiveBayesClassifier = SafeJClass('com.hankcs.hanlp.classification.classifiers.NaiveBayesClassifier')
```

```python
training_data_path = './data/training_data/movie_2'
# training_data_path = './data/training_data/ChnSentiCorp'

def train_classifier(path):
    classifier = NaiveBayesClassifier()
    classifier.train(training_data_path)
    model = classifier.getModel()
    return NaiveBayesClassifier(model)
```

```python
sheet = pd.read_excel(test_data_path)
```

```python
def predict_sentiment(sheet):
    sentiment = []
    for index,row in sheet.iterrows():
        # print(row[0])
        sentiment.append(classifier.classify(row[0]))
    return sentiment
```

```python
def write(sentiment):
    # 在表格中加入新列(结果列)
    sentiment_results = np.array(sentiment)
    sheet.insert(2, 'predict', sentiment_results)
    
    # 设置写回路径
    back_path = './data/results_v3/movie_comments_sentiment.xls'
    sheet.to_excel(back_path) 
```

```python
if __name__ == '__main__':
    classifier = train_classifier(training_data_path)
    sentiment = predict_sentiment(sheet)
    
    write(sentiment)
```
