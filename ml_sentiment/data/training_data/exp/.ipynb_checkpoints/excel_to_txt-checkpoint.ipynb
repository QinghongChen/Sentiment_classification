```python
import pandas as pd
```

```python
excel_path = './douban_comments_clear_round1.xls'


cluster_root_path = './movie_2/'
cluster_name = ['负面/','正面/']
cluster_paths = []
for name in cluster_name:
    cluster_paths.append(cluster_root_path + name)
# print(cluster_paths[1])
```

```python
sheet = pd.read_excel(excel_path, usecols=[1,2,3], header = None)
sheet = sheet.drop(0)
print(sheet)
```

```python
for index,row in sheet.iterrows():
    num = str(row[0])
    txt_name = num+'.txt'
    
    if row[1]<3:
        txt_path = cluster_paths[0]
    else:
        txt_path = cluster_paths[1]
    
    
    text = row[2]
    if len(text)>20:
        file = open(txt_path+txt_name, 'w',encoding='utf-8')
        file.write(text)
    
        
```
