```python
import pandas as pd
```

```python
posData = pd.read_table("./pos.txt", sep="\n",header = None)
negData = pd.read_table("./neg.txt", sep="\n",header = None)
```

```python
for index,row in posData.iterrows():
    if index >= 550:
        posData = posData.drop(index)
```

```python
for index,row in negData.iterrows():
    if index >= 450:
        negData = negData.drop(index)
```
